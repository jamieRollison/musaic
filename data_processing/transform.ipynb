```python
import pandas as pd
import json
```

```python
j = pd.read_json('./final_data_records_jamie.json', orient = 'records')
m = pd.read_json('./final_data_records_melanie.json', orient = 'records')
```

```python
thing = []
for month in range(1, 13):
    record = {}
    data = j[j["month"] == month]
    for feature in [
        "valence",
        "energy",
        "speechiness",
        "danceability",
        "acousticness",
        "tempo",
    ]:
        indices = list(data[feature].nlargest(3).to_dict().keys())
        top_songs = j.iloc[indices]
        record[feature] = top_songs.to_dict('records')
    thing.append(record)

with open("jamie_top.json", "w") as file:
    json.dump(thing, file)
```
