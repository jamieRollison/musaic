```python
def rollingthing(rn: int, df: pd.DataFrame):
    attributes = df.groupby("date")[
    [
        "valence",
        "energy",
        "speechiness",
        "danceability",
        "acousticness",
        "tempo"
    ]].mean()

    roll = attributes.rolling(rn, center=True)
    means = roll.mean()
    dates = np.zeros(means.shape[0], dtype=object)
    dates[: rn // 2] = ""
    dates[-(rn // 2) :] = ""

    for i in range(rn // 2, means.shape[0] - (rn // 2)):
        roll_dates = [means.iloc[j].name for j in range(i - (   rn//2), i + (rn//2)+1)]
        if all([d == roll_dates[0] for d in roll_dates]):
            dates[i] = f"{roll_dates[0][0]}-{roll_dates[0][1]}"
        else:
            dates[i] = f"{roll_dates[0][0]}-{roll_dates[0][1]}:{roll_dates[-1][0]}-{roll_dates[-1][1]}"

    return means.assign(date=dates).iloc[rn // 2 : -(rn //2)]
```

```python
def rollingthing(rn: int, df: pd.DataFrame):
    attributes = df[
        ["valence", "energy", "speechiness", "danceability", "acousticness", "tempo", "date"]
    ]

    # print(attributes)

    roll = attributes.rolling(rn, center=True)
    means = roll.mean()
    dates = np.zeros(means.shape[0], dtype=object)
    dates[: rn // 2] = ""
    dates[-(rn // 2) :] = ""

    for i in range(rn // 2, means.shape[0] - (rn // 2)):
        roll_dates = [
            attributes.iloc[j]['date'] for j in range(i - (rn // 2), i + (rn // 2) + 1)
        ]
        dates[i] = (roll_dates[0], roll_dates[-1])

    return means.assign(date=dates).iloc[rn // 2 : -(rn // 2)]
```

```python
def format_date(dr):
    s, e = dr
    if s == e:
        return f"{s[0]}-{s[1]}"
    return f"{s[0]}-{s[1]}:{e[0]}-{e[1]}"
```

```python
jmeans = rollingthing(20, j)
mmeans = rollingthing(20, m)

mean = jmeans.groupby("date").mean()
jrecords = mean.to_dict('records')
for i, (date, data) in enumerate(mean.iterrows()):
    jrecords[i].update({'date': format_date(date)})

mean = jmeans.groupby("date").mean()
mrecords = mean.to_dict("records")
for i, (date, data) in enumerate(mean.iterrows()):
    mrecords[i].update({"date": format_date(date)})

with open("../src/routes/Visualization/data/rolling_avg.json", "w") as f:
    json.dump([jrecords, mrecords], f)
```
